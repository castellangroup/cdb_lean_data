```python
from xbbg import blp
import pandas as pd
import os
import time
import sys

data = pd.read_csv('stock_crncy_pairs.csv')

print(data)
```

```python
#create a dictionary of country (split ticker by space and take 2nd element) and currency
country_currency = {}

for i in range(len(data)):
    country_currency[data['Ticker'][i].split()[1]] = data['Currency'][i]
```

```python
country_currency

#country currency dictionary to dataframe to csv
country_currency_df = pd.DataFrame(list(country_currency.items()), columns=['Country', 'Currency'])
print(country_currency_df)
```

```python
country_currency_df.to_csv('country_currency.csv', index=False) 
```

```python
tickers = data['Ticker'].values.tolist()
crncy = data['Currency'].values.tolist()
#fields = ["PX_OPEN", "PX_HIGH", "PX_LOW", "PX_LAST", "PX_VOLUME"]
#fields = ["RSI_30D"]
fields = ["PX_LAST"]

for pair in crncy:
    start_date = '2010-01-01'
    end_date = "2024-04-08"

    pair = pair.upper()
    
    #ticker should be USD+currency BGN Curncy, like USDJPY BGN Curncy
    if pair != 'USD':
        ticker = 'USD' + pair + ' BGN Curncy'

        data = blp.bdh(ticker, fields, start_date=start_date, end_date=end_date, Days="A")
        print(data)

        ticker_name = "USD_" + pair

        #path based on folder_dict
        field = fields[0]
        folder = 'fx'

        if folder == 'equity':
            if not os.path.exists(f'equity/{ticker_country}'):
                os.makedirs(f'equity/{ticker_country}')

            data.to_csv(f'equity/{ticker_country}/{ticker_name}.csv', index=True, header=True)
            print('Data saved to ' + f'equity/{ticker_country}/{ticker_name}.csv')

        elif folder == 'fx':
            data.to_csv(f'rates/{ticker_name}.csv', index=True, header=True)
            print('Data saved to ' + f'fx/rates/{ticker_name}.csv')

        else:
            if not os.path.exists(f'factor/{folder}/{ticker_country}'):
                os.makedirs(f'factor/{folder}/{ticker_country}')

            data.to_csv(f'factor/{folder}/{ticker_country}/{ticker_name}.csv', index=True, header=True)
            print('Data saved to ' + f'factor/{folder}/{ticker_country}/{ticker_name}.csv')

        time.sleep(0.05)
```
